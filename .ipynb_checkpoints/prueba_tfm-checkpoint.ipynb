```python
import os
import cv2
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from scipy.spatial import distance_matrix
from scipy.sparse.csgraph import minimum_spanning_tree
from google.colab.patches import cv2_imshow
```

```python
# Obtiene las im√°genes originales
images_or = []
data_dir = './TFM/original_images'
files = os.listdir(data_dir)
for file in files:
  image = cv2.imread(f'{data_dir}/{file}', cv2.IMREAD_GRAYSCALE)
  # image = load_img(f'{data_dir}/{file}')
  images_or.append([image, file])
```

```python
# Rota las im√°genes originales 3 veces (90, 180 y 270)
images_rotated = []
for image, file in images_or:
  images_rotated.append([image, file])
  for i in range(3):
    rotated_image = np.rot90(image, k=i+1)
    images_rotated.append([rotated_image, f"{file}_{i}"])

# Muestra las im√°genes
for i in range(len(images_rotated)):
  plt.imshow(images_rotated[i][0])
  plt.title(images_rotated[i][1])
  plt.axis('off')
  plt.show()
```

```python
# Guarda las nuevas im√°genes en una carpeta nueva
new_data_dir = './TFM/rotated_images'
os.makedirs(new_data_dir, exist_ok=True)
for image, file in images_rotated:
  cv2.imwrite(f'{new_data_dir}/{file}', image)
```

```python
plt.figure(figsize=(20, 20))
images_fin = []
files = os.listdir(new_data_dir)
print(files)
for i, file in enumerate(files):
  # Lee la imagen en escala de grises
  image = cv2.imread(f'{new_data_dir}/{file}', cv2.IMREAD_GRAYSCALE)
  resized_image = cv2.resize(image, (224, 224))

  # Ajusta el contraste usando ecualizaci√≥n de histograma adaptativo
  clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
  imagen_contrastada = clahe.apply(resized_image)

  # Umbralizaci√≥n para convertir la imagen en blanco y negro puro
  _, imagen_umbralizada = cv2.threshold(imagen_contrastada, 180, 255, cv2.THRESH_BINARY)
  
  # Invierte la imagen
  imagen_invertida = cv2.bitwise_not(imagen_umbralizada)

  # Aplica apertura para eliminar peque√±os detalles
  kernel = np.ones((3,3),np.uint8)
  imagen_invertida = cv2.morphologyEx(imagen_invertida, cv2.MORPH_OPEN, kernel)

  # Aplica filtro Gaussiano para eliminar ruido
  imagen_filtrada = cv2.GaussianBlur(imagen_invertida, (5, 5), 0)

  # Detecta de bordes con par√°metros ajustados
  edges = cv2.Canny(imagen_filtrada, threshold1=100, threshold2=200)

  # Encuentra contornos
  contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  # Filtra contornos por tama√±o
  min_area = 5  # Ajusta este valor seg√∫n el tama√±o de los puntos negros
  filtered_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > min_area]

  # Dibuja los contornos detectados
  stars_detected = cv2.cvtColor(imagen_invertida, cv2.COLOR_GRAY2BGR)
  cv2.drawContours(stars_detected, filtered_contours, -1, (0, 255, 0), 2)  

  # Guarda la iamgen en la lista de im√°genes finales
  images_fin.append([imagen_umbralizada, file])
```

```python
# Separa las im√°genes de los target
X_train = [x[0] for x in images_ar]
nombres = [x[1] for x in images_ar]
```

```python
X_train = np.array(X_train)
X_train
```

```python
nombres
```

```python
# Quita la extensi√≥n de los nombres
nombres = [x.split('.')[0] for x in nombres]
nombres
```

```python
# Saca las constelaciones √∫nicas
nombres_unicos = set(nombres)
nombres_unicos
```

```python
# Crea un diccionario con el nombre y su valor
diccionario_nombres = {nombre: i for i, nombre in enumerate(nombres_unicos)}
diccionario_nombres
```

```python
# Cambia el nombre por su valor
y_train = [diccionario_nombres[nombre] for nombre in nombres]
y_train
```

```python
# Convierte las etiquetas en one-hot
y_train = tf.keras.utils.to_categorical(y_train, num_classes=8)
y_train
```

```python
# Crea la red neuronal para las estrellas
model = Sequential([
    # Tama√±o de entrada
    Input((224, 224, 1)),
    # Primera capa
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    # Pasa de dos dimensiones a una
    Flatten(),
    # Capa oculta
    Dense(64, activation='relu'),
    # Capa de salida
    Dense(8, activation='softmax')
])
# Muestra resumen del modelo
model.summary()
```

```python
# Compila el modelo con el optimizador, la funci√≥n de p√©rdida, y la m√©trica
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Entrena el modelo con 10 iteraciones y tama√±o de lote de 8
model.fit(X_train, y_train, batch_size=8, epochs=10)
```
